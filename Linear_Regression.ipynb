# To implement Linear Regression on a simple dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Generate some sample data points to implement Linear Regression on

```python
np.random.seed(0)
x = np.random.rand(1000, 1)
y = 5*x + 3 + np.random.randn(1000, 1) * 0.5 #add some noise to line y=5x+3
```

```python
data=pd.DataFrame(data={"Feature": x.flatten(), "Target":y.flatten()})

data.head()
```

```python
plt.scatter(x, y, color="red")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Dataset")
plt.show()
```

### Now we implement linear regression

```python
def predict(x, w, b):
    return x * w + b

def cost_function(x, y, w, b):
    m = len(y)
    predictions = predict(x, w, b)
    loss = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return loss

# Gradient Descent
def gradient_descent(x, y, w, b, l_r, num_iterations):
    m = len(y)  # Number of data points
    losses = []  # To store loss at each iteration
    for i in range(num_iterations):
        # Compute predictions
        predictions = predict(x, w, b)
        
        dw = (1 / m) * np.sum((predictions - y) * x)
        db = (1 / m) * np.sum(predictions - y)
        
        
        w -= l_r * dw
        b -= l_r * db
        
        loss = cost_function(x, y, w, b)
        losses.append(loss)
        
        print(f"Iteration {i}: Loss = {loss}")
    
    return w, b, losses
```

```python

w = np.random.randn(1)  #initial weights
b = np.random.randn(1)  #initialintercept
learning_rate = 0.1  # Learning rate
num_iterations = 1000  # Number of iterations


# Train the model
w, b, losses = gradient_descent(x, y, w, b, learning_rate, num_iterations)
print(f"Trained weight: {w}")
print(f"Trained bias: {b}")

# Make predictions
y_pred = predict(x, w, b)

# Plot the dataset and the regression line
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue')
plt.plot(x, y_pred, color='red')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression')

plt.subplot(1, 2, 2)
plt.plot(range(num_iterations), losses, color='green')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Gradient Descent Curve')

plt.tight_layout()
plt.show()
```
